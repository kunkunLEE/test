```python
import numpy as np
import matplotlib.pyplot as plt

x = y = np.arange(-1.5, 1.5, 0.01)
x, y = np.meshgrid(x,y)
plt.contour(x, y, x**2 + y**2, [1])     #x**2 + y**2 = 9 的圆形

plt.axis('scaled')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = y = np.arange(-1.5, 1.5, 0.1)
x, y = np.meshgrid(x,y)
plt.contour(x, y, abs(x) + abs(y), [1])     #x**2 + y**2 = 9 的圆形

plt.axis('scaled')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0
x = y = np.arange(-1.5, 1.5, 0.01)
x, y = np.meshgrid(x,y)
plt.contour(x, y, abs(x)**p + abs(y)**p, [2])     #x**2 + y**2 = 9 的圆形

plt.axis('scaled')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

z_list = []
for y in np.linspace(-1.5,1.5,100):
    for x in np.linspace(-1.5,1.5,100):
        z = max(abs(x),abs(y))
        z_list.append(z)    #获得z的数据
z = z_list    
x = np.linspace(-1.5,1.5,100)
y = np.linspace(-1.5,1.5,100)      
#因为z是一维，所以要变成3*3
z = np.mat(z)              
z = np.array(z)
z.shape = (100,100)
plt.figure(figsize=(5,5))
plt.contour(x,y,z,[1])
```
